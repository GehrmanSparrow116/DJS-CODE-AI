```python
import pandas as pd
df=pd.read_csv(r"C:\Users\caash\Downloads\deliveries_updated_mens_ipl_upto_2024 - deliveries_updated_mens_ipl_upto_2024.csv")
```

```python
import plotly.express as px
bb=["J J Bumrah","B Kumar"]
bb_df=df[df["bowler"].isin(bb)]
fig=px.scatter(bb_df,x="dismissal_kind",y="bowler",hover_data=["dismissal_kind"])
fig.show()
```

```python
batters=["V Kohli","RG Sharma","HH Pandya"]
b_df=df[(df["batsman"].isin(batters)) & (df["batsman_runs"].isin([4,6]))]
bo_cnt=b_df.groupby(["batsman","over_ball","batsman_runs"]).size().reset_index(name="count")
fig = px.bar(bo_cnt,x="over_ball",y="count",color="batsman",hover_data=["batsman"])
fig.update_layout(height=600)
fig.show()
```
