```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,MinMaxScaler
import seaborn as sns
df=pd.read_csv(r"C:\Users\caash\Downloads\deliveries_updated_mens_ipl_upto_2024 - deliveries_updated_mens_ipl_upto_2024.csv")
```

```python
top_teams = df["batting_team"].value_counts().nlargest(5).index
zoom_df = df[df["batting_team"].isin(top_teams)]
plt.figure(figsize=(16,8))
sns.violinplot(data=zoom_df,x="batting_team",y="batsman",hue="dismissal_kind")
plt.legend(title='Batting Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.yticks([])
plt.title("Batsman Dismissal Kind")
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.countplot(data=df, x="batsman_runs", hue="batting_team", palette="tab20")
plt.legend(title='Batting Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.title("Frequency of Run Types by Team")
plt.show()
```

```python
plt.figure(figsize=(100,50))
sns.barplot(data=df,x="bowler",y="over_ball",hue="bowling_team",palette="tab20")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks([])
plt.xlabel("Bowlers")
plt.ylabel("Overs")
plt.title("Bowlers Wicket")
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(data=df,x="bowling_team",y="extras")
plt.title("Extras Bowled By Each Team")
plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("Extras")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,MinMaxScaler
import seaborn as sns
df=pd.read_csv(r"C:\Users\caash\Downloads\deliveries_updated_mens_ipl_upto_2024 - deliveries_updated_mens_ipl_upto_2024.csv")
plt.figure(figsize=(16,8))
sns.lineplot(data=df,x="over",y="dismissal_kind",hue="batting_team",palette="colorblind")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
teams=["Mumbai Indians","Chennai Super Kings","Rising Pune Supergiants"]
new_df=df[df["batting_team"].isin(teams)]
dismissal_counts = new_df.groupby(["batting_team", "dismissal_kind"]).size().reset_index(name="count")
sns.lineplot(data=dismissal_counts,x="batting_team",y="count",hue="dismissal_kind",palette="colorblind")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
extra_col=["isWide","isNoBall","Byes","LegByes","Penalty"]
extras_sum=df.groupby("bowling_team")[extra_col].sum()
sns.heatmap(extras_sum,annot=True,cmap="YlGnBu",fmt=".0f")
plt.show()
```
